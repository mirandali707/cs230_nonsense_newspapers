```python
def plot_losses(loss_dict):
    iterations = loss_dict['iter']
    lossG = loss_dict['lossG']
    lossD = loss_dict['lossD']
    
    plt.plot(iterations, lossG, label = "lossG")
    plt.plot(iterations, lossD, label = "lossD")

    plt.xlabel('iteration no.')
    plt.ylabel('loss')
    
    plt.title(f"generator/discriminator loss for scale {loss_dict['scale']}")
    plt.legend()

    plt.show()
```

```python
def plot_all_losses(pkl_filename):
    with open(pkl_filename, 'rb') as f:
        losses = pickle.load(f)
    for i in range(len(losses)):
        plot_losses(losses[i])
```
